<pre style='font-size : 20px'><b>PART 1</b></pre>
<pre style='font-size : 18px'><b>Predicting housing prices with SVM</b></pre>
<pre style='font-size : 15px;'>Firstly let's import necessary packages. </pre>

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
```

<pre style='font-size : 15px'>As alwasys, let's load the data.</pre>

```python
def load_boston_df():
    from sklearn import datasets
    boston_data = datasets.load_boston()
    columns_to_use = ['data','target','feature_names']
    boston_data = dict([(i,boston_data[i]) for i in columns_to_use])
    df = pd.DataFrame(columns=boston_data['feature_names'], data=boston_data['data'])
    df['target'] = boston_data['target']
    return df
```

```python
df = load_boston_df()
```

```python
df.head()
```

<pre style='font-size : 15px'>We created a function <b>process</b> to process, scale and split our data.</pre>

```python
def process(df,seed=5,scale=True):
    train_columns = list(df.columns)
    target_column = train_columns.pop(-1)
    train_set, test_set = train_test_split(df,test_size=0.20, random_state=seed)
    X_train, y_train, X_test, y_test = train_set[train_columns], train_set[target_column],\
    test_set[train_columns], test_set[target_column]
    if scale:
        scaler = MinMaxScaler().fit(X_train)
        X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)
    return X_train, y_train, X_test,y_test, test_set 
```

```python
X_train, y_train, X_test, y_test, test_set = process(df,seed=5
                                          )
```

<pre style='font-size : 15px'>Now let's define our Support Vector Regressor and train it on our data. </pre>

```python
#C - the regularization term, the bigger C - the smaller the regularization effect and vice versa
C = 100
#kernel is a type of kernel used for classifier, could be one of the following : 
#'linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’ or a callable
kernel = 'rbf'
#degree - polynomial degree if kernel = 'poly'
degree = 2
#gamma - controlsthe width of margin
gamma = 1
#feel free to play with all parameters yourself and see the results
```

```python
clf = SVR(C=C,kernel=kernel,degree=degree,gamma=gamma
         )
```

```python
clf.fit(X_train,y_train)
```

```python
y_predicted_train = clf.predict(X_train)
print("Mean squared error on train data : {0}".format(mean_squared_error(y_predicted_train,y_train)))
```

```python
y_predicted_test = clf.predict(X_test)
print("Mean squared error on test data : {0}".format(mean_squared_error(y_predicted_test,y_test)))
```

<pre style='font-size :15px'>Using the specified hyper parameters we got much better results on both train and test data. 
There are many combinations of different hyper parameters of the current model, thus feel free to play with them and see the results!</pre>

<pre style='font-size : 20px'><b>PART 2</b></pre>
<pre style='font-size : 18px'><b>Visualizing predictions</b></pre>

<pre style='font-size :15px'>By visualizing prediction boundaries the one can understand the dependencies between different hyper parameters of classifer. Function <b>plot_rbf</b> plots the decision boundary
of SVR using radial basis kernel function, 
while <b>plot_poly</b> plots it with respect to polynomial kernel function.
Try it yourself.</pre>

```python
sns.set()
def plot_rbf(gamma=1,C=100):
    clf = SVR(C=C,kernel='rbf',gamma=gamma
         )
    clf.fit(X_train,y_train)
    y_predicted_test = clf.predict(X_test)
    mae = mean_squared_error(y_predicted_test,y_test)
    plt.figure(figsize=(15,12))
    plt.title('Predictions on test data with C of {0} ; test mse score of {1} ; gamma of {2}'.format(C,mae, gamma))
    plt.scatter(y_test,test_set['RM'])
    plt.scatter(y_predicted_test, test_set['RM'],color='g', marker='^')
    plt.ylabel('Number of rooms')
    plt.xlabel('Price')
    plt.legend(['Real value','Predicted value'])
```

```python
plot_rbf()
```

```python
plot_rbf(gamma=1,C=0.1)
```

```python
plot_rbf(gamma=10,C=10)
```

```python
plot_rbf(gamma=100,C=0.1)
```

```python
def plot_poly(gamma=1,C=100,degree=2):
    clf = SVR(C=C,kernel='poly',gamma=gamma, degree=degree
         )
    clf.fit(X_train,y_train)
    y_predicted_test = clf.predict(X_test)
    mae = mean_squared_error(y_predicted_test,y_test)
    plt.figure(figsize=(15,12))
    plt.title('Predictions on test data with C of {0} ; test mse score of {1} ; gamma of {2} ; degree of {3}'
              .format(C,mae, gamma, degree))
    plt.scatter(y_test,test_set['RM'])
    plt.scatter(y_predicted_test, test_set['RM'],color='g', marker='^')
    plt.ylabel('Number of rooms')
    plt.xlabel('Price')
    plt.legend(['Real value','Predicted value'])
```

```python
plot_poly()
```

```python
plot_poly(C=1,gamma=5, degree=2)
```

```python
plot_poly(C=100,gamma=0.01, degree=2)
```

<pre style='font-size : 15px'>To sum up, using SVR we got even better results with a test mse of 6, 
you also had a chance to see how the results vary with respect to different hyperparameters.
