<pre style='font-size : 15px;'>Firstly let's import necessary packages. </pre>

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
```

<pre style='font-size : 15px'>Now it's time to load our dataset and split it into subsets.</pre>

```python
def load_boston_df():
    from sklearn import datasets
    boston_data = datasets.load_boston()
    columns_to_use = ['data','target','feature_names']
    boston_data = dict([(i,boston_data[i]) for i in columns_to_use])
    df = pd.DataFrame(columns=boston_data['feature_names'], data=boston_data['data'])
    df['target'] = boston_data['target']
    return df
```

```python
df = load_boston_df()
```

<pre style='font-size : 15px'>For decision trees we don't need to normalize and scale our data, thus we will only split it into different subsets.</pre>

```python
def process(df,seed=5,scale=True):
    train_columns = list(df.columns)
    target_column = train_columns.pop(-1)
    train_set, test_set = train_test_split(df,test_size=0.20, random_state=seed)
    X_train, y_train, X_test, y_test = train_set[train_columns], train_set[target_column],\
    test_set[train_columns], test_set[target_column]
    if scale:
        scaler = MinMaxScaler().fit(X_train)
        X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)
    return X_train, y_train, X_test,y_test 
```

```python
X_train, y_train, X_test,y_test = process(df,scale=False)
```

<pre style='font-size : 15px'>Before creating our classifier, we will set some hyperparameters.</pre>

```python
#max depth of the tree, if None
#then nodes are expanded until all leaves are pure 
#or until all leaves contain less than min_samples_split samples
max_depth = 6
#the minimum number of samples required to split an internal node
min_split = 4
#the number of features to consider when looking for the best split
max_features = 8
#number of samples required to make a leaf
min_samples_leaf = 5
#feel free to play with the parameters shown above and see the result

#random state for obtaining the same results
seed=5

#limits total number of leaves in the node
max_leaf_nodes = 15
```

```python
clf = tree.DecisionTreeRegressor(max_depth=max_depth, min_samples_split=min_split, random_state=seed,
                                 max_features=max_features, min_samples_leaf=min_samples_leaf, 
                                max_leaf_nodes=max_leaf_nodes)
```

```python
clf.fit(X_train, y_train)
```

```python
y_predicted_train = clf.predict(X_train)
print("Mean squared error on train data : {0}".format(mean_squared_error(y_predicted_train,y_train)))
```

```python
y_predicted_test = clf.predict(X_test)
print("Mean squared error on test data : {0}".format(mean_squared_error(y_predicted_test,y_test)))
```

<pre style='font-size : 15px'>We got 16 mse score on test and 17 on train data. You can try to play with parameters to get even better results. Now let's plot our decision tree classifier and see its structure.</pre>

```python
plt.figure(figsize=(15,12))
train_columns = list(df.columns)
target_column = train_columns.pop(-1)
tree.plot_tree(clf,feature_names=train_columns, filled=True)
plt.show()
```

<pre style='font-size : 15px'>From the figure above that's obvious how the tree separates the samples of our data. So far we used only one tree for regression, the next step is to use multiple trees for regression. 
We can think about trees as about humans that are making decisions. 
The bigger number of opinions we have, the better results we'll gain. 
Thus, the next classifier we will use is an ensemble of trees called random forest.</pre>

<pre style='font-size : 20px'><b>PART 2</b></pre>
<pre style='font-size : 18px'><b>Predicting housing prices with RandomForestRegressor</b></pre>

```python
from sklearn import ensemble
```

<pre style='font-size : 15px'>As previously let's set the hyperparameters.</pre>

```python
# number of trees to use
n_estimators = 100
#max depth of the tree, if None
#then nodes are expanded until all leaves are pure 
#or until all leaves contain less than min_samples_split samples
max_depth = 12
#the minimum number of samples required to split an internal node
min_split = 5
#the number of features to consider when looking for the best split
max_features = 8
#number of samples required to make a leaf
min_samples_leaf = 5
#limits total number of leaves in the node
max_leaf_nodes = 31
```

```python
clf = ensemble.RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_split,
                              max_features=max_features,min_samples_leaf=min_samples_leaf,
                                    max_leaf_nodes=max_leaf_nodes)
```

```python
clf.fit(X_train,y_train)
```

```python
y_predicted_train = clf.predict(X_train)
print("Mean squared error on train data : {0}".format(mean_squared_error(y_predicted_train,y_train)))
```

```python
y_predicted_test = clf.predict(X_test)
print("Mean squared error on test data : {0}".format(mean_squared_error(y_predicted_test,y_test)))
```

<pre style='font-size : 15px'>Using ensemble of trees we got even better results.
The last classifier from tree family that we will consider is named GradientBoostingRegressor.</pre>

<pre style='font-size : 20px'><b>PART 3</b></pre>
<pre style='font-size : 18px'><b>Predicting housing prices with GradientBoostingRegressor</b></pre>

```python
# number of trees to use
n_estimators = 10000
#max depth of the tree, if None
#then nodes are expanded until all leaves are pure 
#or until all leaves contain less than min_samples_split samples
max_depth = 3
#the minimum number of samples required to split an internal node
min_split = 2
#the number of features to consider when looking for the best split
max_features = 5
#learning rate to fit the classifier
eta = 0.1

```

```python
clf = ensemble.GradientBoostingRegressor(n_estimators=n_estimators, max_depth=max_depth,min_samples_split=min_split,
                                        max_features=max_features,
                                        learning_rate=eta,random_state=seed,
                                       )
```

```python
clf.fit(X_train,y_train)
```

```python
y_predicted_train = clf.predict(X_train)
print("Mean squared error on train data : {0}".format(mean_squared_error(y_predicted_train,y_train)))
```

```python
y_predicted_test = clf.predict(X_test)
print("Mean squared error on test data : {0}".format(mean_squared_error(y_predicted_test,y_test)))
```
